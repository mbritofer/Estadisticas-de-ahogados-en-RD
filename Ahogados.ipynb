<h1>Estadisticas de ahogados en República Dominicana de 2011 al 2019</h1>

<h2>Importamos Librerias necesarias</h2>

```python
import numpy as np # Algebra Linear
import pandas as pd # Procesamiento de Data, Archivo CSV I/O (ej. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

<h2>Importamos archivo con informacion</h2>

```python
df_data = pd.read_csv(r"C:\\Users\\Manuel Brito\\Documents\\Developments\\Python\\Dominican Republic\\Datos Abiertos\\CADSECI\\Ahogados.csv", encoding='latin-1')
df_data.head()
```

<h2>Cantidad total de ahogados en República Dominicana del 2011 al 2019</h2>

```python
df_data.columns = df_data.columns.str.strip()

df_data = df_data.filter(['Año', 'Mes', 'Sexo', 'Lugar', "Provincia"], axis=1)

cc = pd.DataFrame(df_data, columns=['Año'])

cc['Ahogados'] = 1

ahogados_total = cc.groupby(['Año']).sum().reset_index()

ahogados_total
```

<h2>Grafica tipo barra</h2>

```python

sns.barplot(data=ahogados_total, x="Año", y="Ahogados")
sns.color_palette("rocket", as_cmap=True)
plt.xlabel('Años')
plt.ylabel('Cantidad de ahogados')
plt.title('Cantidad de ahogados en República Dominicana del 2011 al 2019')

```

```python
df_data.columns = df_data.columns.str.strip()

df_data = df_data.filter(['Año', 'Mes', 'Sexo', 'Lugar', "Provincia"], axis=1)

dd = pd.DataFrame(df_data, columns=['Año', 'Lugar'])

dd['Ahogados'] = 1

dd
```

<h2>Lugares de ahogados en República Dominicana del 2011 al 2019</h2>

```python
ahogados = dd.groupby(['Año','Lugar']).sum().reset_index()

ahogados
```

```python
sns.catplot(x = "Año",       # x variable name
            y = "Ahogados",       # y variable name
            hue = "Lugar",  # group variable name
            data = ahogados,     # dataframe to plot
            kind = "bar")

plt.title('Lugares de ahogados en República Dominicana del 2011 al 2019')
```

<h2>Lugares y sexos de ahogados en República Dominicana del 2011 al 2019</h2>

<h3>Segmetación de histogramas por subconjuntos de datos</h3>

```python
ee = pd.DataFrame(df_data, columns=['Año', 'Lugar', 'Sexo'])

sns.set_theme(style="darkgrid")
df = pd.DataFrame(ee, columns=['Año', 'Lugar', 'Sexo'])
sns.displot(
    df, x="Año", col="Lugar", row="Sexo",
    binwidth=1, height=3, facet_kws=dict(margin_titles=True),
)

plt.title('Lugares y sexos de ahogados en República Dominicana del 2011 al 2019')
```

<h2>Sexos de ahogados en República Dominicana del 2011 al 2019</h2>

```python
ff = pd.DataFrame(df_data, columns=['Año', 'Sexo'])

ff['Ahogados'] = 1

ahogados_sex = ff.groupby(['Año','Sexo']).sum().reset_index()

ahogados_sex
```

```python
sns.catplot(x = "Año",       # x variable name
            y = "Ahogados",       # y variable name
            hue = "Sexo",  # group variable name
            data = ahogados_sex,     # dataframe to plot
            kind = "bar")

plt.xlabel('Años')
plt.ylabel('Cantidad de ahogados')
plt.title('Sexos de ahogados en República Dominicana del 2011 al 2019')
```

```python
sns.set_theme(style="whitegrid")

#penguins = pd.DataFrame(df_data, columns=['Año', 'Sexo'])

# Draw a nested barplot by species and sex
g = sns.catplot(
    data=ahogados_sex, kind="bar",
    x="Año", y="Ahogados", hue="Sexo",
    errorbar="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("Años", "Cantidad de ahogados")
g.legend.set_title("Sexo")
plt.title('Sexos de ahogados en República Dominicana del 2011 al 2019')
```

<h2>Ahogados por provincias en República Dominicana del 2011 al 2019</h2>

```python
gg = pd.DataFrame(df_data, columns=['Año', 'Provincia'])

gg['Ahogados'] = 1

#tests_df[(tests_df['grade'] > 10)]

hh = gg[(gg['Provincia'] == 'Santo Domingo')]

provincias = hh.groupby(['Año','Provincia']).sum().reset_index()

provincias
```

```python
sns.catplot(x = "Año",       # x variable name
            y = "Ahogados",       # y variable name
            hue = "Provincia",  # group variable name
            data = provincias,     # dataframe to plot
            kind = "bar")

plt.title('Total de ahogados por provincias en República Dominicana del 2011 al 2019')
```

<h3>Total de ahogados en Santo Domingo del 2011 al 2019</h3>

```python
g = sns.catplot(x="Año", y="Ahogados", kind='bar', palette="rocket", data = provincias)
g.fig.suptitle("Total de ahogados en Santo Domigo del 2011 al 2019", size=20)
g.set_axis_labels("Años", "Nro. de Ahogados", size=16)
plt.subplots_adjust(top=.9)
# extract the matplotlib axes_subplot objects from the FacetGrid
ax = g.facet_axis(0, 0)

# iterate through the axes containers
for c in ax.containers:
    labels = [f'{(v.get_height() ):.0f}' for v in c]  #/ 1000 K
    ax.bar_label(c, labels=labels, label_type='edge', padding=6)
plt.show()
```

```python
provincias_todas = gg.groupby(['Año','Provincia']).sum().reset_index()

g = sns.FacetGrid(provincias_todas, col='Provincia')
g.map(sns.scatterplot, 'Año', 'Ahogados', alpha=0.5)
#g.set(ylim=(750, -10));
```
